```python
### most important players using betweenness centrality
btw_total={}
btw_count={}
included=[]

for i in range(group):
  #print("##########Group##########   ",i+1)
  btw_dict={}
  btw_dict=nx.betweenness_centrality(G[i+1], normalized = True)
  
  for j in range(len(player_list)):
    if player_list[j] in btw_dict.keys():
      if player_list[j]=='n41':
        print('n41= ',btw_dict[player_list[j]])
        
      if player_list[j] in btw_total:
        btw_total[player_list[j]]+=btw_dict[player_list[j]]
        btw_count[player_list[j]]+=1
      else:
        included.append(player_list[j])
        btw_total[player_list[j]]=btw_dict[player_list[j]]
        btw_count[player_list[j]]=1
    

print(btw_total['n41'])
print(btw_count['n41'])

btw_mean={}
for j in range(len(player_list)):
  if player_list[j] in btw_total.keys():
    #btw_mean[player_list[j]]=btw_total[player_list[j]]/btw_count[player_list[j]]
    btw_mean[player_list[j]]=btw_total[player_list[j]]/group

print('No_important player= ',len(btw_mean.keys()))

print(included)
print(len(included))
```

```python
##### Find most important players using Betweenness centrality

def find_max(from_dict):
  max_v=0
  max_k=0
  v=list(from_dict.values())
  k=list(from_dict.keys())
  for i in range(len(k)):
    if max_v<from_dict[k[i]]:
      max_k=k[i]
      max_v=from_dict[k[i]]

  return max_k, max_v

max_k, max_v=find_max(btw_mean)
print("1st most important= ",max_k)
print("1st most value= ",f"{max_v:0.5f}")

print(" ")
btw_mean2=btw_mean.copy()
del btw_mean2[max_k]
max_k2, max_v2=find_max(btw_mean2)
print("2nd most important= ",max_k2)
print("2nd most value= ",f"{max_v2:0.5f}")

print(" ")
btw_mean3=btw_mean2.copy()
del btw_mean3[max_k2]
max_k3, max_v3=find_max(btw_mean3)
print("3rd most important= ",max_k3)
print("3rd most value= ",f"{max_v3:0.5f}")
```

```python
##### Find most important players in the dictionary

def find_max(from_dict):
  max_v=0
  max_k=0
  v=list(from_dict.values())
  k=list(from_dict.keys())
  for i in range(len(k)):
    if max_v<from_dict[k[i]]:
      max_k=k[i]
      max_v=from_dict[k[i]]

  return max_k, max_v
```
